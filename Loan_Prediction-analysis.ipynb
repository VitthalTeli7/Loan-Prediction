```python
def empirical_analysis(col):
    mu = loan_df[col].mean()
    sigma = loan_df[col].std()
    minus1 = mu - sigma ; plus1 = mu + sigma
    minus2 = mu - (2*sigma) ; plus2 = mu + (2*sigma)
    minus3 = mu - (3*sigma) ; plus3 = mu + (3*sigma)
    x1 = loan_df[(loan_df[col] > minus1) & (loan_df[col] < plus1)]
    x2 = loan_df[(loan_df[col] > minus2) & (loan_df[col] < plus2)]
    x3 = loan_df[(loan_df[col] > minus3) & (loan_df[col] < plus3)]
    return(x1,x2,x3)
```

```python
def median_imputed(col):
    q1=np.quantile(loan_df[col],0.25)
    q2=np.quantile(loan_df[col],0.50)
    q3=np.quantile(loan_df[col],0.75)

    Iqr=q3-q1
    ub=q3+(1.5*Iqr)
    lb=q1-(1.5*Iqr)
    con1=loan_df[col]>ub
    con2=loan_df[col]<lb
    outliers_df=loan_df[con1|con2]
    new_values=[]
    median=round(loan_df[col].median(),2)
    for i in loan_df[col].values:
        if i>ub or i<lb:
            new_values.append(median)
        else:
            new_values.append(i)

    loan_df[f'{col}_imputed'] = new_values
```
